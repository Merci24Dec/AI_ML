```python
import theano as th
import theano.tensor as tt
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = tt.fvector('x')
y = tt.fvector('y')
m = th.shared(0.8,'m')
c = th.shared(0.9,'c')
# Regression Equation
yhat = m*x + c
```

```python
# calculate the cost function
cost = tt.mean(tt.sqr(y-yhat))
```

```python
# Gradient Descent Algorithm
LR = 0.03
mn = m - LR*tt.grad(cost,m)
cn = c - LR*tt.grad(cost,c)
```

```python
# create a theano function to train 
train = th.function(inputs=[x,y],outputs=cost,updates=[(m,mn),(c,cn)])
```

```python
area = np.array([1.8,1.6,2.4,3.5,6.5,4.8,5.7,7.2,6.2,3.5,8.1,7.6,4.2])
price= np.array([210,270,260,280,590,520,640,810,590,380,740,690,370])
area = area.astype('float32')
price = price.astype('float32')
```

```python
for i in range(5000):
    error = train(area,price)
    print(error)
```
