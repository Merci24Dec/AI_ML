### Exploratory Analysis
#### Bivariate Analysis

```python
# Numeric v/s Categorical
# CreditScore v/s Exited
plt.figure(figsize=(12,5))
sns.distplot(df.CreditScore[df.Exited==0])
sns.distplot(df.CreditScore[df.Exited==1])
plt.legend(['0','1'])
plt.show()
```

```python
# Numeric v/s Categorical
# Age v/s Exited
plt.figure(figsize=(12,5))
sns.distplot(df.Age[df.Exited==0])
sns.distplot(df.Age[df.Exited==1])
plt.legend(['0','1'])
plt.show()
```

```python
# the above graph shows that majority of customers leaving are from
# old age group, young age customers have less dropout rate
```

```python
# Categorical v/s Categorical
# Geography v/s Exited
plt.figure(figsize=(6,3))
sns.countplot(df.Geography)
plt.show()
plt.figure(figsize=(6,3))
sns.countplot(df.Geography[df.Exited==1])
plt.show()
```

```python
# Categorical v/s Categorical
# Gender v/s Exited
plt.figure(figsize=(6,3))
sns.countplot(df.Gender)
plt.show()
plt.figure(figsize=(6,3))
sns.countplot(df.Gender[df.Exited==1])
plt.show()
```

```python
# Categorical v/s Categorical
# HasCrCard v/s Exited
plt.figure(figsize=(6,3))
sns.countplot(df.HasCrCard)
plt.show()
plt.figure(figsize=(6,3))
sns.countplot(df.HasCrCard[df.Exited==1])
plt.show()
```

### Multivariate Analyssi

```python
# pointplot
# categorical v/s numeric v/s categorical
# x, y, hue >> x = categorical, y = numeric, hue = Categorical

plt.figure(figsize=(12,5))
sns.pointplot(x ='Geography',y='Age',hue='Exited',data=df)
plt.show()
```

```python
# numeric v/s numeric v/s categorical
# creditscore v/s age v/s exited
plt.figure(figsize=(12,5))
sns.scatterplot(x='CreditScore',y='Age',hue='Exited',data=df)
plt.show()
```

```python
# Correlation analysis
cor = df.corr()
# Heatmap for visualizing correlation analysis
plt.figure(figsize=(10,8))
sns.heatmap(cor,annot=True,cmap='coolwarm')
plt.show()
```
